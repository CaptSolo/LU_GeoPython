## NumPy arrays

NumPy arrays are the main way we will use NumPy. They are similar to Python lists, but can have any number of dimensions. They are also much faster than Python lists, and have a number of useful methods.

It is a table of elements (usually numbers), all of the same type, indexed by a tuple of positive integers. In NumPy dimensions are called *axes*.

Key characteristics of NumPy arrays:

- **Homogeneous** - all elements are of the same type
- **Fixed-size** - the size of the array is fixed when it is created
- **Fast** - NumPy arrays are implemented in C, so they are much faster than Python lists
- **Convenient** - NumPy arrays have a number of useful methods


NumPy’s array class is called *ndarray*. It is also known by the alias *array*.

##  When should you use ndarray instead of normal Python data structures such as list?

* NumPy's arrays are more compact than Python lists (about 3-5x difference)
* Faster access for reading/writing items / less Python "bookkeeping"
* Convencience operations: matrix operations/Fourier Transforms/histograms/statistics/filtering/etc
* Spatial locality in memory access patterns results in performance gains notably due to the CPU cache
* Items stored contiguously in memory allow NumPy to take advantage of vectorized instructions of modern CPUs, such as Intel's SSE and AVX, AMD's XOP


* Restriction(uniform data types)

## Most important array attributes

* ndarray.ndim

the number of axes (dimensions) of the array.
* ndarray.shape

the dimensions of the array. This is a tuple of integers indicating the size of the array in each dimension. For a matrix with n rows and m columns, shape will be (n,m). The length of the shape tuple is therefore the number of axes, ndim.
* ndarray.size

the total number of elements of the array. This is equal to the product of the elements of shape.
* ndarray.dtype

an object describing the type of the elements in the array. One can create or specify dtype’s using standard Python types. Additionally NumPy provides types of its own. numpy.int32, numpy.int16, and numpy.float64 are some examples.
* ndarray.itemsize

the size in bytes of each element of the array. For example, an array of elements of type float64 has itemsize 8 (=64/8), while one of type complex32 has itemsize 4 (=32/8). It is equivalent to ndarray.dtype.itemsize.
* ndarray.data

the buffer containing the actual elements of the array. Normally, we won’t need to use this attribute because we will access the elements in an array using indexing facilities.

### Practical example - calculating the value of Pi

Monte-Carlo method:
- https://www.geeksforgeeks.org/estimating-value-pi-using-monte-carlo/


![](https://upload.wikimedia.org/wikipedia/commons/thumb/d/d4/Pi_monte_carlo_all.gif/406px-Pi_monte_carlo_all.gif)

```python
import random as rnd
```

```python
# without Numpy

def approximate_pi(num_points):

    inside = 0
    
    for count in range(num_points):
        x = rnd.random()
        y = rnd.random()

        # Calculate how many points fall inside the unit circle
        if x**2 + y**2 <= 1:
            inside += 1

    # Calculate the approximation of pi
    pi_approximation = (inside / num_points) * 4
    
    return pi_approximation
```

```python
%%time

# Example with 10 million points
num_points = 10_000_000
pi_estimate = approximate_pi(num_points)
print(f"Approximation of pi: {pi_estimate}")
```

```python
# with Numpy

import numpy as np

def approximate_pi_numpy(num_points):
    # Generate random x and y coordinates
    x = np.random.rand(num_points)
    y = np.random.rand(num_points)

    # Calculate how many points fall inside the unit circle
    inside_circle = np.sum(x**2 + y**2 <= 1)

    # Calculate the approximation of pi
    pi_approximation = (inside_circle / num_points) * 4
    return pi_approximation
```

```python
%%time 

# Example with 10 million points
num_points = 10_000_000
pi_estimate = approximate_pi_numpy(num_points)
print(f"Approximation of pi: {pi_estimate}")
```

Leibniz formula for Pi:
- https://en.wikipedia.org/wiki/Leibniz_formula_for_%CF%80

$$\pi = 4 \sum_{n=0}^{\infty} \frac{(-1)^n}{2n + 1}$$


```python
# Leibniz formula for Pi

def leibniz_pi(num_terms):
    n = np.arange(num_terms)
    terms = (-1)**n / (2 * n + 1)
    return 4 * np.sum(terms)
```

```python
%%time

print("Leibniz Pi approximation:", leibniz_pi(10_000_000))
```

## Solving Systems of Linear Equations

Numpy makes it easy to solve systems of linear equations, which is common in engineering, physics, and economics.

Example:

$$3x + 4y = 7$$
$$2x − y = 3$$

```python
# Define coefficients matrix A and constants vector B

A = np.array([[3, 4], [2, -1]])
B = np.array([7, 3])

# Solve for the variables

solution = np.linalg.solve(A, B)

print("Solution for x and y:", solution)
```

## Broadcasting 

* The term broadcasting describes how numpy treats arrays with different shapes during arithmetic operations. Subject to certain constraints, the smaller array is “broadcast” across the larger array so that they have compatible shapes

* usually leads to efficient algorithm implementations.

```python
e = np.arange(6)
f = np.arange(3)*10  #same as np.arange(0,30,10)
```

### General Broadcasting Rules
#### When operating on two arrays, NumPy compares their shapes element-wise. It starts with the trailing dimensions, and works its way forward. Two dimensions are compatible when

* they are equal, or
* one of them is 1

## Example (network analysis)

```python
# Adjacency matrix for a social network (5 people)
# 1 means friendship, 0 means no connection

network = np.array([
    [0, 1, 1, 0, 0],  # Person 0's connections
    [1, 0, 1, 1, 0],  # Person 1's connections
    [1, 1, 0, 1, 1],  # Person 2's connections
    [0, 1, 1, 0, 1],  # Person 3's connections
    [0, 0, 1, 1, 0]   # Person 4's connections
])
```

```python
# direct friends
direct_friends = network.sum(axis=1)
print("Direct friends per person:", direct_friends)
```

```python
# Matrix multiplication = 2nd-degree connections
second_degree = network @ network
print("\n2nd degree connections:\n", second_degree)
```

```python
# Most connected person
most_connected = np.argmax(direct_friends)
print(f"\nMost connected person: Person {most_connected}")
```
